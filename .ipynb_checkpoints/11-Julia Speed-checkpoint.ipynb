# Speed of Julia

Consider the **sum** function `sum(a)`, which computes
$$
sum(a) = \sum_{i=1}^n a_i
$$
where $n$ is the length of `a`

```julia
arr = rand(10 ^ 7)
```

```julia
sum(arr)
```

### Benchmarking a few ways in a few languages
Julia has a `BenchmarkTools.jl` package for easy and accurate benchmarking

```julia
@time sum(arr)
```

The `@time` macro can yield noisy results, so it is not the best choice for benchmarking

```julia
using Pkg
Pkg.add("BenchmarkTools")
```

```julia
using BenchmarkTools
using Statistics
```

#### C

```julia
using Libdl
C_code = """
#include <stddef.h>
double c_sum(size_t n, double *X) {
    double s = 0.0;
    for(size_t i = 0; i < n; ++i)
        s += X[i];
    return s;
}
"""

const Clib = tempname() # Make a temporary file

# compile to a shared library by piping C code to gcc
open(`gcc -fPIC -O3 -msse3 -xc -shared -ffast-math -o $(Clib * "." * Libdl.dlext) -`, "w") do f
    print(f, C_code)
end

c_sum(X::Array{Float64}) = ccall(("c_sum", Clib), Float64, (Csize_t, Ptr{Float64}), length(X), X)
```

```julia
c_sum(arr)
```

```julia
c_sum(arr) ≈ sum(arr)
```

```julia
≈ # alias for the `isapprox` function
```

```julia
?isapprox
```

```julia
c_bench = @benchmark c_sum(arr)
```

```julia
println("C: Fastest time was $(minimum(c_bench.times) / 1e6) ms")
```

```julia
d = Dict()
d["C"] = minimum(c_bench.times) / 1e6
d
```

```julia
using Plots
gr()
```

```julia
t = c_bench.times / 1e6
m, σ = minimum(t), std(t)

histogram(t, bins=500,
    xlim=(m-.01, m+σ),
    xlabel="milliseconds", ylabel="count", label=""
)
```

#### Python

```julia
# Pkg.add("PyCall")
using PyCall
```

```julia
py_sum = pybuiltin("sum")
```

```julia
py_sum(arr)
```

```julia
py_sum(arr) ≈ sum(arr)
```

```julia
py_bench = @benchmark py_sum(arr) 
```

```julia
d["Python built-in"] = minimum(py_bench.times) / 1e6
d
```

#### NumPy

```julia
Pkg.add("Conda")
using Conda
```

```julia
Conda.add("numpy")
```

```julia
np_sum = pyimport("numpy")["sum"]
np_bench = @benchmark np_sum(arr)
```

```julia
d["NumPy"] = minimum(np_bench.times) / 1e6
d
```
